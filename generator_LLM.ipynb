# Data Processing

```python
import string
```

```python
#remove brackets [] from Lyrics
data['Lyrics'] = data['Lyrics'].apply(lambda x: x.replace('[','').replace(']','').replace("'",'').replace(",",'').
                                      replace('"','').replace('(','').replace(')','') if isinstance(x, str) else x)
```

```python
lyricsConcat = data['Lyrics'].str.cat(sep=' ')
```
